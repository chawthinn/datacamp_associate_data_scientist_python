Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
import pandas as pd

schools = pd.read_csv("schools.csv")

schools.head()
```

Question: Which NYC schools have the best math results?

* The best math results are at least 80% of the *maximum possible score of 800* for math.
* Save your results in a pandas DataFrame called best_math_schools, including "school_name" and "average_math" columns, sorted by "average_math" in descending order.

```python
schools.info()
```

```python
# Subset by math scores at least 80% of 800
best_math_schools = schools[schools['average_math'] >= 800*(.80)]

best_math_schools.info()
```

```python
best_math_schools['average_math']
```

```python
# Save as a DataFrame
best_math_schools = best_math_schools[['school_name', 'average_math']].sort_values(by='average_math', ascending=False)
```

```python
best_math_schools
```

Question: What are the top 10 performing schools based on the combined SAT scores?

* Save your results as a pandas DataFrame called top_10_schools containing the "school_name" and a new column named "total_SAT", with results ordered by "total_SAT" in descending order ("total_SAT" being the sum of math, reading, and writing scores).

```python
# Create new column to store total scores
schools['total_SAT'] = (
    schools['average_math'] + 
    schools['average_reading'] + 
    schools['average_writing']
)

schools.info()
```

```python
schools.head()
```

```python
# Create a new df with cols 'school_name' and 'total_score' and sort by SAT
sorted_df = schools[['school_name', 'total_SAT']].sort_values(by='total_SAT', ascending=False)

sorted_df.head()
```

```python
# Select only top 10 schools
top_10_schools = sorted_df.head(10)

print(top_10_schools)
```

Question: Which single borough has the largest standard deviation in the combined SAT score?

* Save your results as a pandas DataFrame called largest_std_dev.
* The DataFrame should contain one row, with:
    * "borough" - the name of the NYC borough with the largest standard deviation of "total_SAT".
    * "num_schools" - the number of schools in the borough.
    * "average_SAT" - the mean of "total_SAT".
    * "std_SAT" - the standard deviation of "total_SAT".
* Round all numeric values to two decimal places.

```python
# Group schools by borough
grouped_by_borough = schools.groupby('borough')

grouped_by_borough.head()
```

```python
# Calculate the standard deviation of 'total_score' of each borough
std_by_borough = grouped_by_borough['total_SAT'].std()

std_by_borough.sort_values(ascending=False)
```

```python
# Find the borough with largest std
largest_std_borough = std_by_borough.idxmax()

largest_std_borough
```

Answer: 'Manhattan' has the largest standard deviation in the combined SAT score.

```python
# Select all columns for borough with largest std
filtered_borough_data = schools[schools['borough'] == largest_std_borough]

# Calculate the number of schools, average SAT, and standard deviation for that borough
num_schools = filtered_borough_data['borough'].count()
average_SAT = filtered_borough_data['total_SAT'].mean()
std_SAT = filtered_borough_data['total_SAT'].std()

# Round the numeric values to two decimal places
num_schools = round(num_schools, 2)
average_SAT = round(average_SAT, 2)
std_SAT = round(std_SAT, 2)

# Create largest_std_dev dataframe
largest_std_dev = pd.DataFrame({
    'borough': [largest_std_borough],
    'num_schools': [num_schools],
    'average_SAT': [average_SAT],
    'std_SAT': [std_SAT]
})

# Display the result
print(largest_std_dev)
```
