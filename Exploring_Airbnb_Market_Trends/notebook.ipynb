Welcome to New York City, one of the most-visited cities in the world. There are many Airbnb listings in New York City to meet the high demand for temporary lodging for travelers, which can be anywhere between a few nights to many months. In this project, we will take a closer look at the New York Airbnb market by combining data from multiple file types like `.csv`, `.tsv`, and `.xlsx`.

Recall that **CSV**, **TSV**, and **Excel** files are three common formats for storing data. 
Three files containing data on 2019 Airbnb listings are available to you:

**data/airbnb_price.csv**
This is a CSV file containing data on Airbnb listing prices and locations.
- **`listing_id`**: unique identifier of listing
- **`price`**: nightly listing price in USD
- **`nbhood_full`**: name of borough and neighborhood where listing is located

**data/airbnb_room_type.xlsx**
This is an Excel file containing data on Airbnb listing descriptions and room types.
- **`listing_id`**: unique identifier of listing
- **`description`**: listing description
- **`room_type`**: Airbnb has three types of rooms: shared rooms, private rooms, and entire homes/apartments

**data/airbnb_last_review.tsv**
This is a TSV file containing data on Airbnb host names and review dates.
- **`listing_id`**: unique identifier of listing
- **`host_name`**: name of listing host
- **`last_review`**: date when the listing was last reviewed

```python
# Import necessary packages
import pandas as pd
import numpy as np

price_df = pd.read_csv("data/airbnb_price.csv")
room_type_df = pd.read_excel("data/airbnb_room_type.xlsx")
last_review_df = pd.read_csv("data/airbnb_last_review.tsv",sep = '\t')
```

```python
print(price_df.head())
print(room_type_df.head())
print(last_review_df.head())
```

```python
print(price_df.info())
print(room_type_df.info())
print(last_review_df.info())
```

```python
combined_df = pd.merge(price_df, room_type_df, on='listing_id')
```

```python
combined_df = pd.merge(combined_df, last_review_df, on='listing_id')
```

```python
print(combined_df.info())
```

What are the dates of the earliest and most recent reviews? 

Store these values as two separate variables with your preferred names.

```python
# Convert last_review to datetime and extract date
combined_df['last_review'] = pd.to_datetime(combined_df['last_review']).dt.date

print(combined_df['last_review'])
```

```python
first_reviewed = combined_df['last_review'].min()
last_reviewed = combined_df['last_review'].max()

print("The earliest review was on", first_reviewed)
print("The most recent review was on", last_reviewed)
```

How many of the listings are private rooms? Save this into any variable.

```python
print(combined_df['room_type'].value_counts())
```

```python
# Transform the values of room_type column to lowercase
combined_df['room_type'] = combined_df['room_type'].str.lower()
```

```python
# Verify the changes
print(combined_df['room_type'].value_counts())
```

```python
# Find the count of private room listing
nb_private_rooms = combined_df[combined_df['room_type'] == 'private room'].shape[0]

print(f"There are {nb_private_rooms} of the listings that are private rooms.")
```

What is the average listing price? 

Round to the nearest two decimal places and save into a variable.

```python
print(combined_df['price'].unique())
```

```python
# Remove the dollars from prices and convert to int
combined_df['price'] = combined_df['price'].str.replace(' dollars', '').astype(int)

print(combined_df['price'].head())
```

```python
# Find average listing price rounded to 2 decimals
avg_price = combined_df['price'].mean().round(2)

print(f"The average listing price is {avg_price}.")
```

```python
review_dates = pd.DataFrame(data=[[first_reviewed, last_reviewed, nb_private_rooms, avg_price]], 
                           columns=['first_reviewed', 'last_reviewed', 'nb_private_rooms', 'avg_price'])
```
